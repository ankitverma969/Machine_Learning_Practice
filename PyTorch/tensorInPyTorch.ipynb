Creating a Tensor

```python
#usig empty tensor
x = torch.empty(2,3)  
# x = torch.empty(size=(2,3))  # another way to create empty tensor
# where 2 is number of rows and 3 is number of columns
print(x)
```

```python
# Check the Type and Shape of the tensor
print(f"Type of x : {x.dtype}")
print(f"Shape of x : {x.shape}")
```

```python
#Using Zeros tensor
x = torch.zeros(2,3)
print(x)
```

```python
# Using Ones tensor
x = torch.ones(2,3)
print(x)
```

```python
# Using Random tensor
x = torch.rand(2,3)
print(x)
```

```python
# Use Seed for reproducibility
torch.manual_seed(42)
x = torch.rand(2,3)
print(x)
```

```python
torch.manual_seed(42)
x = torch.rand(2,3)
print(x)
```

```python
# Using Tensor
x = torch.tensor([[1,2,3],[4,5,6]])
print(x)
```

```python
# Other way to create tensor

# 1. arange - like python's range but returns a tensor
x = torch.arange(0,10,2)  # start, end, step
print(x)

# 2. linspace - returns evenly spaced numbers over a specified interval
x = torch.linspace(0,1,5)  # start, end, number of
print(x)

# 3. eye - returns a 2-D tensor with ones on the diagonal and zeros elsewhere
x = torch.eye(5)  # size of the square matrix
print(x)

# 4. diag - returns a diagonal tensor
x = torch.diag(torch.ones(5))  # size of the square matrix
print(x)

# 5. full - returns a tensor filled with a specified value
x = torch.full((2,3), 7)  # size and value
print(x)
```

Tesor Shapes

```python
x = torch.tensor([[1,2,3],[4,5,6]])
print(x)
print(x.dtype)
print(x.shape)
```

```python
# Making a tensor with same sape as another tensor using empty_like, zeros_like, ones_like functions
x = torch.empty_like(x)
print(x)
```

```python
# torch.zeros_like(x)
# torch.ones_like(x)
print(torch.zeros_like(x))
print(torch.ones_like(x))
```

Tensor Data Types

```python
# find the data type of the tensor
print(x.dtype)
```

```python
# Assigning the data type of the tensor
x = torch.tensor([1.0,2.0,3.0], dtype=torch.int32)
print(x)
print(x.dtype)
```

```python
torch.tensor([1,2,3], dtype=torch.float32)
print(x)
print(x.dtype)
```

```python
# change the data type of given data using .to() method
x = x.to(torch.float32)
print(x)
```

Mathematical Operations

```python
x = torch.rand(2,2)
x
```

```python
# Addition
x+2
```

```python
# Subtraction
x-2
print(x)

# Multiplication
x*2
print(x)

# Division
x/2
print(x)
```

```python
# Abs
x = torch.tensor([-1,-2,3])
print(x)
print(torch.abs(x))
```

```python
# Negative
torch.neg(x)
```

```python
# round
x = torch.tensor([1.2, 2.7, 3.1])
print(x)
print(torch.round(x))
```

```python
# Ceil 
x = torch.tensor([1.2, 2.7, 3.1])
print(x)
print(torch.ceil(x))
```

```python
# FLoor
x = torch.tensor([1.2, 2.7, 3.1])
print(x)
print(torch.floor(x))
```

```python
# Clamp
x = torch.tensor([-1.2, 0.5, 2.7, 3.1])
print(x)    
print(torch.clamp(x, min=0.0, max=2.0))
```
