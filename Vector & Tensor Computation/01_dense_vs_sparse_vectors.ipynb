Dense Vector Representation

If most values are non-zero → store everything

If most values are zero → store only useful values

```python
import numpy as np
```

```python
#Dense vector
v_dense = np.array([3,0,0,0,5,0,0])
```

```python
print(v_dense)
print("Length: ", len(v_dense))
```

Sparse Vector Representation

Store only non-zero values and their indices.

```python
from scipy.sparse import csr_matrix

#Sparse vector using CSR format

v_sparse = csr_matrix([3,0,0,0,5,0,0])
print(v_sparse)
```

Comparing Dense vs Sparse (DSA View)

| Property       | Dense         | Sparse      |
| -------------- | ------------- | ----------- |
| Space          | O(n)          | O(k)        |
| Cache friendly | ✅ Yes         | ❌ No        |
| GPU friendly   | ✅ Yes         | ❌ Limited   |
| Best for       | Deep learning | NLP, graphs |
