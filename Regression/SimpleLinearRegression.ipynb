Simple Linear Regression From Scratch (No sklearn)

```python
# -----------------------------------------------------------
# Simple Linear Regression FROM SCRATCH (No sklearn used)
# -----------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
```

```python
# ---------------------------
# Step 1: Create simple data
# ---------------------------
# Relationship: y = 3x + 5 (with some noise)
np.random.seed(0)
X = np.linspace(0, 20, 30)
noise = np.random.randn(30) * 3
y = 3 * X + 5 + noise
```

```python
# ---------------------------
# Step 2: Calculate slope (m) and intercept (b)
# ---------------------------
# Formula:
# m = Σ((x - mean_x)(y - mean_y)) / Σ((x - mean_x)^2)
# b = mean_y - m * mean_x

mean_x = np.mean(X)
mean_y = np.mean(y)

numerator = np.sum((X - mean_x) * (y - mean_y))
denominator = np.sum((X - mean_x)**2)

m = numerator / denominator
b = mean_y - m * mean_x

print("Slope (m):", m)
print("Intercept (b):", b)
```

```python
# ---------------------------
# Step 3: Make predictions
# ---------------------------
y_pred = m * X + b
```

```python
# ---------------------------
# Step 4: Plot results
# ---------------------------
plt.scatter(X, y, label="Data Points")
plt.plot(X, y_pred, color="red", label="Best Fit Line (Calculated)")
plt.title("Simple Linear Regression (From Scratch)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
