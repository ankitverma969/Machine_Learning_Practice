Manhattan Distance

```python
# Manhattan Distance is another way to measure the distance between two points.
# Instead of straight-line distance (like Euclidean), it measures distance along grid paths, like moving through city blocks.

# It is widely used in algorithms such as K-Nearest Neighbors and K-Means Clustering.
```

1. Manhattan Distance From Scratch

```python
def manhattan_distance(point1, point2):
    distance = 0
    for i in range(len(point1)):
        distance += abs(point1[i] - point2[i])
    return distance

# Example
p = [2,3]
q = [5,7]

print("Manhattan Distance:", manhattan_distance(p,q))
```

Using NumPy (Inbuilt)

```python
import numpy as np

point1 = np.array([2,3])
point2 = np.array([5,7])

manhattan_distance = np.sum(np.abs(point1 - point2))

print("Manhattan Distance:", manhattan_distance)
```

Using SciPy Inbuilt Function

```python
from scipy.spatial.distance import cityblock

p = [2,3]
q = [5,7]

distance = cityblock(p,q)

print("Manhattan Distance:", distance)
```

| Feature | Euclidean Distance | Manhattan Distance       |     |   |
| ------- | ------------------ | ------------------------ | --- | - |
| Formula | √Σ(p−q)²           | Σ                        | p−q |   |
| Path    | Straight line      | Grid path                |     |   |
| Shape   | Circle             | Diamond                  |     |   |
| Used in | KNN, clustering    | KNN, grid-based problems |     |   |
