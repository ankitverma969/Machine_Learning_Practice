```python
# tfidf_manual_calculation.py

import math
from collections import Counter

# -----------------------------
# Step 1: Documents
# -----------------------------
documents = [
    "The Cat Sat on Mat",
    "The Dog Played in Park",
    "The Cat and Dogs are great pets"
]

# -----------------------------
# Step 2: Preprocessing
# -----------------------------
stopwords = {"the", "on", "in", "and", "are"}

def preprocess(text):
    text = text.lower()
    words = text.split()
    words = [w for w in words if w not in stopwords]
    return words

processed_docs = [preprocess(doc) for doc in documents]

print("Processed Documents:")
for i, doc in enumerate(processed_docs, 1):
    print(f"Doc{i}: {doc}")

# -----------------------------
# Step 3: Vocabulary
# -----------------------------
vocab = sorted(set(word for doc in processed_docs for word in doc))
print("\nVocabulary:", vocab)

# -----------------------------
# Step 4: Term Frequency (TF)
# -----------------------------
def compute_tf(doc):
    tf_dict = {}
    total_words = len(doc)
    word_count = Counter(doc)
    
    for word in vocab:
        tf_dict[word] = word_count[word] / total_words
    
    return tf_dict

tf_list = [compute_tf(doc) for doc in processed_docs]

# -----------------------------
# Step 5: Document Frequency (DF)
# -----------------------------
df = {}
N = len(processed_docs)

for word in vocab:
    df[word] = sum(1 for doc in processed_docs if word in doc)

print("\nDocument Frequency:", df)

# -----------------------------
# Step 6: Inverse Document Frequency (IDF)
# -----------------------------
idf = {}

for word in vocab:
    idf[word] = math.log(N / df[word])

print("\nIDF Values:")
for k, v in idf.items():
    print(f"{k}: {round(v, 3)}")

# -----------------------------
# Step 7: TF-IDF
# -----------------------------
tfidf_list = []

for tf_doc in tf_list:
    tfidf_doc = {}
    for word in vocab:
        tfidf_doc[word] = tf_doc[word] * idf[word]
    tfidf_list.append(tfidf_doc)

# -----------------------------
# Step 8: Display TF-IDF
# -----------------------------
print("\nTF-IDF Matrix:\n")

for i, doc in enumerate(tfidf_list, 1):
    print(f"Doc{i}:")
    for word, value in doc.items():
        if value > 0:
            print(f"  {word}: {round(value, 3)}")
    print()
```
