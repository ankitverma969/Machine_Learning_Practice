Euclidean Distance in Python

```python
# From Scratch and Using Inbuilt Functions

import numpy as np
import pandas as pd
import math
```

```python
def euclidean_distance(point1, point2):
    sum_squared_distance = 0

    for i in range(len(point1)):
        sum_squared_distance += (point1[i] - point2[i])**2
    
    distance = sum_squared_distance ** 0.5
    return distance

# Example usage
point_a = [1, 2, 3]
point_b = [4, 5, 6]
distance = euclidean_distance(point_a, point_b)
print(f"The Euclidean distance between {point_a} and {point_b} is: {distance}")
```

Euclidean Distance Using NumPy (Inbuilt)

```python
p = np.array([2,3])
q = np.array([5,7])

distance = np.linalg.norm(p - q)

print("Euclidean Distance (NumPy):", distance)
```

Using SciPy Inbuilt Function

```python
from scipy.spatial.distance import euclidean

p = [2,3]
q = [5,7]

distance = euclidean(p,q)

print("Euclidean Distance (SciPy):", distance)
```
